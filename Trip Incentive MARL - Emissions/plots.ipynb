```python
# Sample data: Each budget has two lists (TTT and emissions) with episode-wise values
# Replace these with your actual data
budgets = [
    "Unlimited",
    "100e3",
    "50e3",
    "10e3",
    "SUMO Dynamic UE no",
]  # Example budgets
ttt_data = [
    [
        extract_last_50_mean(TTTs),
        extract_last_50_mean(TTTs1),
        extract_last_50_mean(TTTs_05_05),
    ],
    [
        extract_last_50_mean(TTTs100_1),
        extract_last_50_mean(TTTs100_2),
        extract_last_50_mean(TTTs100_3),
    ],
    [
        extract_last_50_mean(TTTs50_1),
        extract_last_50_mean(TTTs50_2),
        extract_last_50_mean(TTTs50_3),
    ],
    [
        extract_last_50_mean(TTTs10_1),
        extract_last_50_mean(TTTs10_2),
        extract_last_50_mean(TTTs10_3),
    ],
    [extract_last_50_mean(TTTs_dua)],
]
emissions_data = [
    [
        extract_last_50_mean(emissions_total),
        extract_last_50_mean(emissions_total1),
        extract_last_50_mean(emissions_total_05_05),
    ],
    [
        extract_last_50_mean(emissions_total100_1),
        extract_last_50_mean(emissions_total100_2),
        extract_last_50_mean(emissions_total100_3),
    ],
    [
        extract_last_50_mean(emissions_total50_1),
        extract_last_50_mean(emissions_total50_2),
        extract_last_50_mean(emissions_total50_3),
    ],
    [
        extract_last_50_mean(emissions_total10_1),
        extract_last_50_mean(emissions_total10_2),
        extract_last_50_mean(emissions_total10_3),
    ],
    [extract_last_50_mean(emissions_total_dua) / 1000],
]
# Function to extract the mean of the last 50 episodes (for the sake of the example, we just use the data directly)
# Since we have only a few points for each budget, we'll just directly work with them


# Prepare the plot
plt.figure(figsize=(10, 6))

# Define a set of colors for each budget
colors = ["blue", "green", "red", "purple", "orange"]

# Plot Pareto fronts for each budget
for i, budget in enumerate(budgets):
    # Extract TTT and emissions data for the specific budget
    ttt = ttt_data[i]
    emissions = emissions_data[i]

    # Plot the Pareto front for this specific budget
    plt.scatter(ttt, emissions, color=colors[i], label=f"{budget} budget", zorder=5)

    # Sort TTT and emissions for this budget (if we want to create a Pareto front line)
    sorted_indices = np.argsort(ttt)
    plt.plot(
        np.array(ttt)[sorted_indices],
        np.array(emissions)[sorted_indices],
        color=colors[i],
        marker="o",
        linestyle="-",
        zorder=2,
    )

# Labels and title
plt.xlabel("TTT [s]")
plt.ylabel("CO2 Emissions [kg]")
plt.title("Pareto Front for different budgets")
# Legend
plt.legend()
plt.savefig("multiple_pareto_front.pdf", format="pdf", bbox_inches="tight")

plt.show()
```
